```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV data
file_path = 'results_SHAP_tunning.csv'
data = pd.read_csv(file_path)

# Debug: Print the first few rows of the data
print(data.head())

# Separate data for insertion and deletion
insertion_data = data[data['auc_evaluation_version'] == 'insertion']
deletion_data = data[data['auc_evaluation_version'] == 'deletion']

# Debug: Print the first few rows of the filtered data
print("Insertion Data:")
print(insertion_data.head())
print("Deletion Data:")
print(deletion_data.head())

# Define a custom color palette
custom_palette = sns.color_palette(["red", "green", "blue", "orange", "purple"])

# Plot for insertion
plt.figure(figsize=(6, 3))
sns.lineplot(x='threshold', y='mean_auc', hue='option', data=insertion_data, marker='o', palette=custom_palette)
plt.ylim(81, 81.5)
plt.title('Insertion')
plt.xlabel('Threshold')
plt.ylabel('Mean AUC')
plt.legend(title='Option')
plt.tight_layout()
plt.savefig("insertion_plot.svg", format="svg")
plt.show()

# Plot for deletion
plt.figure(figsize=(6, 3))
sns.lineplot(x='threshold', y='mean_auc', hue='option', data=deletion_data, marker='o', palette=custom_palette)
plt.ylim(22.2, 22.7)
plt.title('Deletion')
plt.xlabel('Threshold')
plt.ylabel('Mean AUC')
plt.legend(title='Option')
plt.tight_layout()
plt.savefig("deletion_plot.svg", format="svg")
plt.show()

# Plot average time taken for each threshold
plt.figure(figsize=(6, 3))
sns.lineplot(x='threshold', y='avg_time', hue='option', data=data, marker='o', palette=custom_palette)
plt.title('Average Time vs Threshold')
plt.xlabel('Threshold')
plt.ylabel('Average Time (seconds)')
plt.legend(title='Option')
plt.tight_layout()
plt.savefig("average_time_plot.svg", format="svg")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV data
file_path = 'results_SHAP_tunning.csv'
data = pd.read_csv(file_path)

# Debug: Print the first few rows of the data
print(data.head())

# Separate data for insertion and deletion
insertion_data = data[data['auc_evaluation_version'] == 'insertion']
deletion_data = data[data['auc_evaluation_version'] == 'deletion']

# Debug: Print the first few rows of the filtered data
print("Insertion Data:")
print(insertion_data.head())
print("Deletion Data:")
print(deletion_data.head())

# Define a custom color palette
custom_palette = sns.color_palette(["red", "green", "blue", "orange", "purple"])

# Plot AUC vs. Threshold for insertion and deletion
plt.figure(figsize=(12, 6))

# Plot for insertion
plt.subplot(1, 2, 1)
sns.lineplot(x='threshold', y='mean_auc', hue='option', data=insertion_data, marker='o', palette=custom_palette)
plt.ylim(81, 81.5)
plt.title('Insertion')
plt.xlabel('Threshold')
plt.ylabel('Mean AUC')
plt.legend(title='Option')

# Plot for deletion
plt.subplot(1, 2, 2)
sns.lineplot(x='threshold', y='mean_auc', hue='option', data=deletion_data, marker='o', palette=custom_palette)
plt.ylim(22.2, 22.7)
plt.title('Deletion')
plt.xlabel('Threshold')
plt.ylabel('Mean AUC')
plt.legend(title='Option')

plt.tight_layout()
plt.show()

# Plot average time taken for each threshold
plt.figure(figsize=(6, 3))
sns.lineplot(x='threshold', y='avg_time', hue='option', data=data, marker='o', palette=custom_palette)
plt.title('Average Time vs Threshold')
plt.xlabel('Threshold')
plt.ylabel('Average Time (seconds)')
plt.legend(title='Option')
plt.show()
# Save as .svg
# plt.savefig("tunning_time.svg", format="svg")
```

# Plotting number of features vs number of subsets

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Number of features
num = 13
n = np.arange(0, num)

# Number of possible coalitions for SHAP calculation
coalitions = 2 ** n

# Set the seaborn style
sns.set(style="whitegrid")

# Plotting the graph
plt.figure(figsize=(10, 6))
sns.lineplot(x=n, y=coalitions, marker='o', label='Number of Coalitions')
plt.axhline(y=100, color='r', linestyle='--', label='y = 100')

# Adding labels and title
plt.xlabel('Number of Features (n)')
plt.ylabel('Number of Possible Coalitions')
plt.title('Number of Possible Coalitions for SHAP Calculation')
plt.legend()

# Set x-ticks in increments of 1
plt.xticks(np.arange(0, num+1, 1))

# Show the plot
plt.grid(True)

# Save the plot as an SVG file
plt.savefig("coalitions_plot.svg", format="svg")

plt.show()
```
